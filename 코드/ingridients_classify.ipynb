```python
cnt = 0
name = df['칵테일명'][cnt]
print(name)
```

```python
df['ingrid_classification']=""
df.head()
```

```python
while( cnt != df['칵테일명'].count()):
    while (df['칵테일명'][cnt] == name):
        if(df['Category'][cnt] =='Ingredient'):
            if(
                'Hot'in df['재료명'][cnt] or 
                'Boiling'in df['재료명'][cnt]
              ):
                df['ingrid_classification'][cnt] = 0
                cnt += 1
            elif(
                'Lemon'in df['재료명'][cnt] or 
                'lemon'in df['재료명'][cnt] or
                'Orange'in df['재료명'][cnt] or
                'Lime'in df['재료명'][cnt] or
                'Fresca'in df['재료명'][cnt]
              ):
                df['ingrid_classification'][cnt] = 1
                cnt += 1
            elif(
                'Grenadine'in df['재료명'][cnt] or 
                'Pinapple'in df['재료명'][cnt] or
                'Cranberr'in df['재료명'][cnt] or
                'Cherry'in df['재료명'][cnt] or
                'Apple'in df['재료명'][cnt] or
                'Peach'in df['재료명'][cnt] or
                'Banana'in df['재료명'][cnt] or
                'Grape'in df['재료명'][cnt] or
                'Coconut'in df['재료명'][cnt] or
                'Passion'in df['재료명'][cnt] or
                'Strawberr'in df['재료명'][cnt] or
                'Mango'in df['재료명'][cnt] or
                'Apricot'in df['재료명'][cnt] or
                'Berr'in df['재료명'][cnt] or
                'Blackberr'in df['재료명'][cnt] or
                'Cantaloupe'in df['재료명'][cnt] or
                'Guava'in df['재료명'][cnt] or
                'Kiwi'in df['재료명'][cnt] or
                'Papaya'in df['재료명'][cnt] or
                'Mint' in df['재료명'][cnt] or
                'Fruit' in df['재료명'][cnt] or
                'Blackcurrant' in df['재료명'][cnt] or
                'Raspberry'in df['재료명'][cnt]
              ):
                df['ingrid_classification'][cnt] = 2
                cnt += 1
            elif(
                'Nectar'in df['재료명'][cnt] or 
                'Agave'in df['재료명'][cnt] or
                'Caramel'in df['재료명'][cnt] or
                'Orgeat'in df['재료명'][cnt] or
                'Cinnamon'in df['재료명'][cnt] or
                'Hazlenut'in df['재료명'][cnt] or
                'Nutmeg'in df['재료명'][cnt] or
                'Sweetened'in df['재료명'][cnt] or
                'Almond'in df['재료명'][cnt] or
                'Cordial'in df['재료명'][cnt] or
                'Maple'in df['재료명'][cnt]
              ):
                df['ingrid_classification'][cnt] = 3
                cnt += 1
            elif(
                'Cream'in df['재료명'][cnt] or 
                'Milk'in df['재료명'][cnt] or
                'Coffee'in df['재료명'][cnt] or
                'Egg'in df['재료명'][cnt] or
                'Chocolate'in df['재료명'][cnt] or
                'Ice-cream'in df['재료명'][cnt] or
                'Yoghurt'in df['재료명'][cnt] or
                'Cocoa'in df['재료명'][cnt] or
                'Tomato'in df['재료명'][cnt] or
                'Espresso'in df['재료명'][cnt] or
                'Yolk'in df['재료명'][cnt] or
                'Vanila'in df['재료명'][cnt]
              ):
                df['ingrid_classification'][cnt] = 4
                cnt += 1
            elif(
                'Soda'in df['재료명'][cnt] or 
                'Coca-cola'in df['재료명'][cnt] or
                '7-up'in df['재료명'][cnt] or
                'Cider'in df['재료명'][cnt] or
                'Tonic'in df['재료명'][cnt] or
                'Sprite'in df['재료명'][cnt] or
                'Schweppes'in df['재료명'][cnt] or
                'Cola'in df['재료명'][cnt] or
                'Pepsi'in df['재료명'][cnt] or
                'Dr. Pepper'in df['재료명'][cnt] or
                'Carbonated'in df['재료명'][cnt]
              ):
                df['ingrid_classification'][cnt] = 6
                cnt += 1
            elif(
                'Ginger'in df['재료명'][cnt] or 
                'Cloves'in df['재료명'][cnt] or
                'Pepper'in df['재료명'][cnt] or
                'Allspice'in df['재료명'][cnt] or
                'Cardamom'in df['재료명'][cnt] or
                'Anis 'in df['재료명'][cnt]
              ):
                df['ingrid_classification'][cnt] = 5
                cnt += 1
            else:
                cnt += 1
        else:
            cnt+=1
        if(cnt == df['칵테일명'].count()):
            break
    if(cnt == df['칵테일명'].count()):
        break
    else:
        name = df['칵테일명'][cnt]
```
