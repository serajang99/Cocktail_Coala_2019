```python
cnt2 = 0 
while( cnt != df['칵테일명'].count()):
    sum = 0
    sum2 = 0
    while (df['칵테일명'][cnt] == name):
        sum += df['oz화개수'][cnt]
        cnt += 1 
        if(cnt == df['칵테일명'].count()):
            break
    for cnt2 in range(cnt2,cnt,1):
        sum2 += df['oz화개수'][cnt2]*df['ABV'][cnt2]
    
    if(df['alcohol'][cnt2]>=0 and df['alcohol'][cnt2]<6):
        df['alcohol_classification'][cnt2] = 0
    elif(df['alcohol'][cnt2]>=6 and df['alcohol'][cnt2]<15):
        df['alcohol_classification'][cnt2] = 1
    elif(df['alcohol'][cnt2]>=15 and df['alcohol'][cnt2]<25):
        df['alcohol_classification'][cnt2] = 2
    elif(df['alcohol'][cnt2]>=25):
        df['alcohol_classification'][cnt2] = 3
    
    cnt2=cnt
    if(cnt == df['칵테일명'].count()):
        break
    else:
        name = df['칵테일명'][cnt]
```
